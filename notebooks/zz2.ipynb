```python
from letta_client import Letta
from pprint import pprint

client = Letta(base_url="http://localhost:8283")
```

```python
agents = list(client.agents.list())

print(f"=== Total Agents Created: {len(agents)} ===")
for a in agents:
    print(f"- {a.name} (ID: {a.id})")
```

```python
agent = client.agents.create(
    name="qwen-agent20260327",
    llm_config={
        "model": "qwen/qwen3.5-35b-a3b",
        "model_endpoint_type": "openai",
        "model_endpoint": "http://100.81.194.12:1234/v1",
        "model_wrapper": None,
        "context_window": 8192
    },
    memory_blocks=[
        {"label": "persona", "value": "You are a helpful assistant powered by LM Studio."},
        {"label": "human", "value": "The user is testing Letta with an external model."},
    ]
)

print(f"Created agent: {agent.id}")
```

```python
agent = client.agents.create(
    name="qwen-agent20260327",
    model="lmstudio_openai/qwen/qwen3.5-35b-a3b",
    context_window_limit=8192,
    memory_blocks=[
        {"label": "persona", "value": "You are a helpful assistant powered by LM Studio."},
        {"label": "human", "value": "The user is testing Letta with an external model."},
    ]
)

print(f"Created agent: {agent.id}")
```

```python
# Send a message
response = client.agents.messages.create(
    agent_id=agent.id,
    input="What do you know about me?"
)

for message in response.messages:
    print(message)
```
