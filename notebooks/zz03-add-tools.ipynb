```python
from letta_client import Letta
from utils.agent_info import get_tool_id_by_name
from utils.letta_tools import DefaultTools

client = Letta(base_url="http://localhost:8283")
```

```python
agent_id = "agent-c425212e-1a63-401d-bff4-048293392e7a"
agent = client.agents.retrieve(
    agent_id=agent_id,
)
```

```python
# 1. Use the auto-completing registry to pick the exact tool name (e.g. Conversation Search)
# No need to memorize the name or spell it out as a string!
tool_name = DefaultTools.ARCHIVAL_MEMORY_INSERT
```

```python
# 2. Dynamically fetch that tool's Letta UUID so you never have to hardcode "tool-abcd-123..."
my_tool_id = get_tool_id_by_name(tool_name, client=client)
```

```python
# 3. Safely attach it to the agent
client.agents.tools.attach(
    agent_id=agent.id, 
    tool_id=my_tool_id
)
```
